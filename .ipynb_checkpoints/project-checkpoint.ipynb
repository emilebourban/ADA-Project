# Women's presence on Wikipedia: A comprehensive analysis on the evolution of women's recognition over time, country and ethnicity.

## Abstract
In the recent century, in a lot of countries, women have finally gained more rights and we have been continuously progressing towards a more equal society. Our goal in this project is to try to highlight the evolution of social gender inequalities in different domains through the ages. Our interest is on the achievements and recognition of the work done by women in different fields as far back as the data goes. We will use the data from Wikipedia and gather data on number of women referenced, their contribution to their domain, and other parameters and compare it with the same data for men. We will also try to find how the country of origin and the time of acquisition of the rights has an significant impact.


## Research questions
Can we accurately use wikipedia database to show the gender inequalities through time?

Are there any evidence that the men/women equality is reached?

What are the domains in which there is more/less equality? Does this change according to regions/country/language?

What is the effect of women empowering events: Right to work, Sufragettes, War effort, ...?

Since some countries have delayed women's rights, is the evolution similar, in term of timeframe, extent?

## Datasets
For this project, we used the [wikidata](https://dumps.wikimedia.org/wikidatawiki/entities/) dataset which repertoriates the data from every Wikipedia article. It is stored in the form of a JSON data dump where the top level structure is organised as folows:
 - id
 - type
 - labels
 - descriptions
 - aliases
 - claims
 - sitelinks
 - lastrevid
 - modified
 
Here we are only interested in keeping some of the properties for every article, mainly the `instance of`, `sex or gender`, `field of work`, `country of origin` and `date of birth`. From this we will make our analysis and conclude.
 
 
To have an easier time visualizing the data at first, we used the [WHGI](http://whgi.wmflabs.org/). This is based on the wikidata JSON file and updated weekly. This data is stored in the form of a `.csv` and contains information about the ethnicity, field of work and times of death/birth divided into different genders. For our analysis, we are only interested in comparing data for the two most frequent genders: male and female.
 
As an extra comparaison tool, we also used the [UN population data](https://population.un.org/wpp/Download/Standard/Population/) that stores a lot of info about the population of each country in gender and age groups. This data only goes back to 1960 but can still be usefull for recentcomparaison.

Here we highlight that we can show some interesting data regarding gender equality using wikidata. After this first analysis we think that it can be used to some extent to highlight gender inequalities through time and space. A caveat however is that we don't have reliable data for the more recent years: even if our society was truely equal in all domains from tomorrow, it would take at least decades before it would show on wikipedia, since people will only get a page after being famous which would happen later in life. Another problem we might encounter is with the temporal resolution of our analysis. Since we can only use the dates of birth or death of a person and not the date when their fame came, we have a high variability between birth, fame and death. Because of this, wikidata seems to be limited for the analysis of single point events repercussions, e.g. right to vote, suffragettes, etc.). One way to overcome this is by a comparison between countries, but then other variables can influence the analysis. Thus, although it can be a usefull tool for our analysis and give us some significant insights regarding the problem of gender inequalities, this datasets still has limits.

## Evolution of Wiki articles per gender over time

```python
# Extract the date of birth and death for every year and gender
dob = d['dateofbirth'].rename(columns={'Unnamed: 0': 'year'}).set_index('year')
dod = d['dateofdeath'].rename(columns={'Unnamed: 0': 'year'}).set_index('year')
dob.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dod.index, dod['male'].apply(lambda r: r/dod.male.max()), 
         dod.index, dod['female'].apply(lambda r: r/dod.female.max()))

plt.legend(['deaths males', 'deaths females'])
plt.xlabel('Year'); plt.ylabel('Normalized article count')
plt.title('Normalized article count by gender (year of death)');
```

**Observation:**

From these plots, we can see that the amount of articles for both men and women increases with time, this is expected since there are more known people closer to us in time and they are better known by whoever is currently living and writting articles about them.
Still, at every point in time, there are always more articles about men than women.

When we look at the the normalized data (for both birth and death years), we can see that the increase in articles about women always starts later than the one for males.

Finally, an intersting point is that for the mot recent years, the amount of articles for men is decreasing after the 1940s; while there are more and more articles for women as time progresses. 

Potentially in a few decades those could reach an equilibrium.

## Overall number of articles with respect to total world population

Here we will try to correct the number of articles with the total world paopulation

```python
wpop = pd.read_csv('wpop.tsv', sep='\t')
```

```python
def clean_year(s):
    """
    The data had a strange format, this helps clean the values for the years
    """
    try:
        if s.endswith('BC') :
            return -int(s[:-3])
        elif s.endswith('AD'):
            return int(s[:-3])
        elif (s[1] == ','):
            # Comma separation for thousands
            return int(s[0])*1e3 + int(s[2:5])
        else:
            return int(s)
    except:
        return None
```

```python
wpop.head(40)
wpop['year'] = wpop['year'].apply(lambda r: clean_year(r))

wpop['world'] = wpop['world'].apply(lambda r: clean_year(r))
```

```python
# World population over time
plt.plot(wpop.year, wpop.world)
```

```python
# Article count normalized by population
dob_wpop = wpop.merge(dob, on='year', how='inner')
plt.plot(dob_wpop.year, dob_wpop.female/dob_wpop.world, 
         dob_wpop.year, dob_wpop.male/dob_wpop.world)
plt.xlabel(''); plt.ylabel('')
plt.title('');
```

**Observation:**

By normalizing with the total wolrd pop, we see that the number of articles per living human decreases in the recent decades. Also while the articles about men has a sharp decline the female articles seem to stay more or less constant. This states that the articles about women is proportial to the world population but for the men they were overexpressed before and now tend to an equilibrium with a lower propartionality constant.

## Occupations per gender

```python
def merge_others(df):
    '''Merges the non-binary gender columns from df into a single one'''
    cols = df.columns.values.tolist()[1:]
    df.fillna(0, inplace=True)
    if 'male' in cols:
        cols.remove('male')
    if 'female' in cols:
        cols.remove('female')
    df['others'] = df[cols].sum(axis=1, numeric_only =True)
    df.drop(cols, inplace=True, axis=1)
    return
```

```python
# we aggregate the sum of non-binary genders in column others
for df in d:
    merge_others(d[df])
```

```python
d['occupation'].rename(columns = {'Unnamed: 0':'Occupation'}, inplace=True)
```

```python
#creating new columns for total number of people and ratio of women
merge_others(d['culture'])
d['occupation']['Total'] = d['occupation'][['female', 'male', 'others']].sum(axis=1, numeric_only =True)
d['occupation']['ratio_w'] = d['occupation']['female'].div(d['occupation']['Total'])

#only taking top 1000 since there are a lot of professions with only a few people doing it so the numbers would not be significant

occupations_1000 = d['occupation'].sort_values('Total', ascending=False).head(1000)

occupations_1000.sort_values('ratio_w', ascending=False, inplace=True)
bins = pd.cut(occupations_1000['ratio_w'],100)
occupations_1000['ratio_w'].hist(bins=50)
plt.xlabel('ratio of women')
plt.ylabel('number of occupations')
plt.title('Histogram of the number of occupation for each ratio of woman (1 is only women)')
```

This graph shows that from a general point of view , a vast majority of occupations of people on wikipedia are largely dominated by men. Since this is taken over all times, it seem to reflect the reality accurately: through time, most occupations succeptible to fame were mostly reserved to men. A further analysis could be made to investigate if this histogram would be similar for different time period.

```python
def return_category(x, dic):
    '''Returns the category of the occupation. If unknown, returns others'''
    for key, vals in dic.items():
        if x in vals:
            return key
       
    return 'others'
```

```python
#we create a column containg the category of each occupation
occupations_200['category'] = occupations_200['Occupation'].apply((lambda x: return_category(x, categories)))
```

```python
# group the occupations by category and caculate the ratio of woman for each
categories = occupations_200.groupby(occupations_200['category']).sum()
categories['ratio_w'] = categories['female'].div(categories['Total'])
categories.sort_values('ratio_w', inplace=True)
```
